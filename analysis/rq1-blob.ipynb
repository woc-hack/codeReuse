# 1-2 Years

```r
x0=read.table("../data/blobs/0y.14",sep=";");
x1=read.table("../data/blobs/1y.14",sep=";");
x2=read.table("../data/blobs/2y.14",sep=";");
```

```r
x1=x1[x1$t<1619874000,] # May 1, 2021
x2=x2[x2$t<1588338000,] # May 1, 2020
x=list(x0,x1,x2)
names(x)=c("All","1y","2y")
for(i in 1:3) {
    names(x[[i]])=c("s","l","t","c")
}
```

What fraction of blobs are copied?

```r
for(i in 1:3) {
    print(c(names(x)[i],sum(x[[i]]$c)/dim(x[[i]])[1]))
}
```

Are copied blobs larger or smaller?

```r
for(i in 1:3) {
    print(names(x)[i])
    print(t.test(log(x[[i]][x[[i]]$c==1,"s"]),log(x[[i]][x[[i]]$c==0,"s"])))
}
```

Are older blobs more likely to be copied?

```r
for(i in 1:3) {
    print(names(x)[i])
    print(t.test(x[[i]][x[[i]]$c==1,"t"],x[[i]][x[[i]]$c==0,"t"]))
}
```

```r
for(i in 2:3) {
    print(names(x)[i]); 
    print(summary(glm(x[[i]]$c~x[[i]]$t+log(x[[i]]$s))))
}
```

Does the propensity to copy vary with language?

```r
for(i in 1:3){
    print(names(x)[i])
    ty=names(table(x[[i]]$l)[table(x[[i]]$l)>100000])
    for(l in ty){ 
        print(c(l,sum(x[[i]][x[[i]]$l==l,"c"])/dim(x[[i]][x[[i]]$l==l,])[1]));
    }
}
```

Does language affect copied vs not copied blob size?

```r
for(i in 1:3){
    print(names(x)[i])
    ty=names(table(x[[i]]$l)[table(x[[i]]$l)>100000])
    for(l in ty){ 
        test=t.test(log(x[[i]][x[[i]]$l==l&x[[i]]$c==1,"s"]),log(x[[i]][x[[i]]$l==l&x[[i]]$c==0,"s"]))
        print(c(l,test$statistic,test$p.value));
    }
}
```

Does language affect the copied vs not copied time(are older blobs more likely to be copied?)?

```r
for(i in 1:3){
    print(names(x)[i])
    ty=names(table(x[[i]]$l)[table(x[[i]]$l)>100000])
    for(l in ty){ 
        test=t.test(x[[i]][x[[i]]$l==l&x[[i]]$c==1,"t"],x[[i]][x[[i]]$l==l&x[[i]]$c==0,"t"])
        print(c(l,test$statistic,test$p.value));
    }
}
```

```r
for(i in 2:3) {
    print(names(x)[i]); 
    ty=names(table(x[[i]]$l)[table(x[[i]]$l)>100000])
    x[[i]]$ll=as.factor(ifelse(x[[i]]$l %in% ty, as.character(x[[i]]$l), "Other"))
    print(summary(glm(x[[i]]$c~x[[i]]$t+log(x[[i]]$s)+x[[i]]$ll), family=binomial))
}
```

```r
for(i in 2:3) {
    print(names(x)[i]); 
    print(summary(glm(x[[i]]$c~log(x[[i]]$s)+x[[i]]$ll), family=binomial))
}
```

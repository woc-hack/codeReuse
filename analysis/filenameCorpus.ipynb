```python
import re
import operator
from collections import Counter
```

```python
path = "../data/fileCorp.0"
with open(path, "r", encoding="ISO-8859-1") as f:
    text = f.read()
text = str.lower(text)
text = re.sub('[/.]', ' ', text)
```

```python
words = re.split('\s+', text)
wf = Counter(words)
wfs = sorted(wf.items(), key = operator.itemgetter(1), reverse=True)
```
